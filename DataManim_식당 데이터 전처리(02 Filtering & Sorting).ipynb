```python
import pandas as pd
url='https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df=pd.read_csv(url)
print(df)
```

```python
#quantity컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하라
#내오답 df.loc['quantity'==3].head(5)           --> df['quantitiy']
df.loc[df['quantity']==3].head(5)
```

```python
#quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라
df[['quantity','item_price']]
```

```python
#item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라
# $2.39 -> 2.39 로 new_price에 숫자로 저장하기
df['new_price']=df['item_price'].str[1:].astype('float')
a = df['new_price'].head()
print(a)
```

```python
#new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라

#내 오답
#df['a']=df[df['new_price']<=5]                --->          loc 사용할 것
#s=df['a'].sum()                        ------> 개수 구하려면 sum 이 아니라 len 사용
#print(s)


#다시풀어보자
#s=len(df.loc[df['new_price']<=5])                     #-----> 조건 주는 방식 df.loc[df.new_price<=5]
#print(s)

#최종 정답
n=len(df.loc[df.new_price<=5])
print(n)
```

```python
#item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하고 index 값을 초기화 하여라
c=df[df.item_name=='Chicken Salad Bowl']
print(c)
```

```python
#index 값을 초기화
c.reset_index(drop=True)
```

```python
#new_price값이 9 이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임을 추출하라
#내 오답
#df[df.new_price<=9 & df.item_name=='Chicken Salad Bowl']        -----> 각 조건을 ()로 감싸주기

#정답
df[(df.new_price<=9)&(df.item_name=='Chicken Salad Bowl')]
```

```python
#df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화 하여라
df.sort_values('new_price').reset_index(drop=True)                           #sort_values , ascending이 기본값
```

```python
#df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라

#보고도 틀린 내오답
# df[:,::2]                      -> iloc 

#최종정답
df.iloc[:,::2]
```

```python
#df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하여라
df.sort_values('new_price',ascending=False).reset_index(drop=True)                #-----> descending=True는 없음 / ascending=False라고해야
```
